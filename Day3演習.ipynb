### Adam

1-2. <font color="Red">Adamã‚’ç”¨ã„ãŸOptimizerã®ã‚¯ãƒ©ã‚¹ Adam ã‚’å®Œæˆã•ã›ã¦ãã ã•ã„ã€‚</font><br>
Adamã®ç‰¹å¾´ã¨ã—ã¦ãƒã‚¤ãƒ‘ãƒ¼ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã®ãƒã‚¤ã‚¢ã‚¹è£œæ­£(åã‚Šã®è£œæ­£)ãŒè¡Œã‚ã‚Œã‚‹ã“ã¨ãŒæŒ™ã’ã‚‰ã‚Œã¾ã™ã€‚æ›¸ç±ã€ã‚¼ãƒ­ã‹ã‚‰ä½œã‚‹DeepLearningã€ã®é…å¸ƒã‚³ãƒ¼ãƒ‰ã§ã¯ç°¡æ˜“ç‰ˆã®ãŸã‚ã€ãƒã‚¤ã‚¢ã‚¹è£œæ­£ã‚’çµ„ã¿è¾¼ã‚“ã§ã„ã¾ã›ã‚“ã€‚ã“ã®å•é¡Œã§ã¯ãƒã‚¤ã‚¢ã‚¹è£œæ­£ã‚’çµ„ã¿è¾¼ã‚“ã å®Œæˆå½¢ã®Adamã‚³ãƒ¼ãƒ‰ã«ã¤ã„ã¦å›ç­”ã—ã¦ãã ã•ã„ã€‚ã¾ãŸã€éå»ã®Eè³‡æ ¼è©¦é¨“ã§ã¯ã“ã¡ã‚‰ã®å®Œæˆå½¢ã®Adamã‚³ãƒ¼ãƒ‰ãŒå‡ºé¡Œã•ã‚Œã¦ã„ã¾ã™ã€‚

```python
class Adam:

    def __init__(self, lr=0.001, beta1=0.9, beta2=0.999):
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.iter = 0
        self.m = None
        self.v = None
        
    def update(self, params, grads):
        if self.m is None:
            self.m, self.v = {}, {}
            for key, val in params.items():
                self.m[key] = np.zeros_like(val)
                self.v[key] = np.zeros_like(val)
        
        self.iter += 1
        
        for key in params.keys():
            ###### å•1-2-1 ######
            self.m[key] = self.beta1 * self.m[key] + (1 - self.beta1) * grads[key]
            ###### å•1-2-2 ######
            self.v[key] = self.beta2 * self.v[key] + (1 - self.beta2) * grads[key] * grads[key]
            ###### å•1-2-3 ###### 
            #è¬›ç¾©ã§ã¯tã‚’å°ã•ãªå€¤ã¨è¨€ã£ã¦ã„ãŸã€‚ç‰¹ã«å¼•æ•°ãŒãªã„ã‹ã‚‰è‡ªåˆ†ã§é©å½“ãªå®šæ•°ã‚’æ±ºã‚ã‚‹ï¼Ÿå‘¼ã³å‡ºã™å´ã§å®šç¾©ã™ã‚‹ï¼Ÿ
            #m_unbias = self.m[key] / (1 - self.beta1 ** t)
            m_unbias = self.m[key] / (1 - self.beta1 ** self.iter)
            ###### å•1-2-4 ######
            #åŒã˜ãtã¯å®šæ•°ï¼Ÿ
            #v_unbias = self.v[key] / (1 -self.beta2 ** t)
            v_unbias = self.v[key] / (1 -self.beta2 ** self.iter)
            params[key] -= self.lr * m_unbias / (np.sqrt(v_unbias) + 1e-7)
```

## ã‚³ã‚¹ãƒˆé–¢æ•°

å¤šã‚¯ãƒ©ã‚¹åˆ†é¡å•é¡Œãªã®ã§ã€ã‚¯ãƒ­ã‚¹ã‚¨ãƒ³ãƒˆãƒ­ãƒ”ãƒ¼ã‚’ã‚³ã‚¹ãƒˆé–¢æ•°ã—ã¦ç”¨ã„ã¾ã™ã€‚

```python
def cross_entropy_error(y, t):
    if y.ndim == 1:
        t = t.reshape(1, t.size)
        y = y.reshape(1, y.size)
    if t.size == y.size:
        t = t.argmax(axis=1)
    batch_size = y.shape[0]
    return -np.sum(np.log(y[np.arange(batch_size), t] + 10e-7)) / batch_size
```

```python
def softmax(x):
    x = x.T
    _x = x - np.max(x, axis=0)
    _x = np.exp(_x) / np.sum(np.exp(_x), axis=0)
    return _x.T
```

## ãƒãƒƒãƒˆãƒ¯ãƒ¼ã‚¯ã®å®Ÿè£…

ã¾ãšã¯ãƒãƒƒãƒæ­£è¦åŒ–ã‚’å…¥ã‚Œãªã„æ™®é€šã®ä¸‰å±¤ãƒ‹ãƒ¥ãƒ¼ãƒ©ãƒ«ãƒãƒƒãƒˆãƒ¯ãƒ¼ã‚¯ã‚’å®Ÿè£…ã—ã¾ã™ã€‚å•é¡Œã«ã¯ãªã£ã¦ã„ã¾ã›ã‚“ãŒã€day1ã®å¾©ç¿’ã‚‚å…¼ã­ã¦ã‚³ãƒ¼ãƒ‰ã‚’èª­ã¿ç†è§£ã—ã¦ãŠã„ã¦ãã ã•ã„ã€‚

```python
class mnistMultiLayerNet:
    """
    layer0: 784 æ¬¡å…ƒã®å…¥åŠ›
    â†“ w1, b1 ã§ç·šå½¢çµåˆ
    â†“ relu ã§æ´»æ€§åŒ–
    layer1: 100 æ¬¡å…ƒã®éš ã‚Œå±¤
    â†“ w2, b2 ã§ç·šå½¢çµåˆ
    â†“ relu ã§æ´»æ€§åŒ–
    layer2: 100 æ¬¡å…ƒã®éš ã‚Œå±¤
    â†“ w3, b3 ã§ç·šå½¢çµåˆ
    â†“ relu ã§æ´»æ€§åŒ–
    layer3: 100 æ¬¡å…ƒã®éš ã‚Œå±¤
    â†“ w4, b4 ã§ç·šå½¢çµåˆ
    â†“ relu ã§æ´»æ€§åŒ–
    layer4: 100 æ¬¡å…ƒã®éš ã‚Œå±¤
    â†“ w5, b5 ã§ç·šå½¢çµåˆ
    layer5: 10 æ¬¡å…ƒã®å‡ºåŠ›å±¤
    """
    def __init__(self):
        self.input_size = 784
        self.output_size = 10
        self.hidden_size_list = [100, 100, 100, 100]
        self.all_size_list = [self.input_size] + self.hidden_size_list + [self.output_size]
        self.hidden_layer_num = len(self.hidden_size_list)
        self.weight_decay_lambda =0
        self.params = {}
        self.layers = {}
        self.grads = {}

        # é‡ã¿ã¨ãƒã‚¤ã‚¢ã‚¹ã®åˆæœŸåŒ–
        for idx in range(1, len(self.all_size_list)):
            self.params['w' + str(idx)] = np.random.randn(self.all_size_list[idx-1], self.all_size_list[idx]) * 0.085
            self.params['b' + str(idx)] = np.zeros(self.all_size_list[idx], dtype=float)

        
    def forward(self, x):
        relu = lambda x : np.maximum(0, x)  # æ´»æ€§åŒ–é–¢æ•°ã¨ã—ã¦ ReLU ã‚’ä½¿ç”¨
        self.layers['layer0'] = x
        for idx in range(1, len(self.all_size_list) - 1):
            w = self.params['w' + str(idx)]
            b = self.params['b' + str(idx)]
            x = self.layers['layer' + str(idx - 1)]
            self.layers['layer' + str(idx)] = relu(np.dot(x, w) + b)
        idx = len(self.all_size_list) - 1
        w = self.params['w' + str(idx)]
        b = self.params['b' + str(idx)]
        x = self.layers['layer' + str(idx - 1)]
        self.layers['layer' + str(idx)] = softmax(np.dot(x, w) + b)
        
        return self.layers['layer' + str(idx)]
        

    def loss(self, y, t):
        return cross_entropy_error(y, t)
    
    def backward(self, t, y):
        delta = (y - t) / t.shape[0]
        self.grads['b5'] = np.sum(delta, axis=0)
        self.grads['w5'] = np.dot(self.layers['layer4'].transpose(), delta)
        # èª¤å·®é€†ä¼æ’­
        for idx in range(4, 0, -1):
            delta = np.dot(delta, self.params['w' + str(idx + 1)].transpose())
            delta = delta *  (self.layers['layer' + str(idx)] > 0)
            self.grads['b' + str(idx)] = np.sum(delta, axis=0)
            self.grads['w' + str(idx)] = np.dot(self.layers['layer'+str(idx - 1)].transpose(), delta)
        return self.grads
```

## ãƒãƒƒãƒæ­£è¦åŒ–ã‚’ç”¨ã„ã‚‹ãƒãƒƒãƒˆãƒ¯ãƒ¼ã‚¯

å„å±¤ã«ã¤ã„ã¦ã€é‡ã¿ã‚’æ›ã‘ã¦è¶³ã—åˆã‚ã›ãŸå¾Œãƒãƒƒãƒæ­£è¦åŒ–ã‚’è¡Œã†ã€‚

2. <font color="Red">ãƒãƒƒãƒæ­£è¦åŒ–ã‚’ç”¨ã„ãŸãƒ‹ãƒ¥ãƒ¼ãƒ©ãƒ«ãƒãƒƒãƒˆãƒ¯ãƒ¼ã‚¯ã‚’å®Œæˆã•ã›ã¦ãã ã•ã„ã€‚</font>

  ãƒãƒƒãƒæ­£è¦åŒ–ã®é †ä¼æ’­ã¯ä»¥ä¸‹ã®å¼ã«å¾“ã£ã¦å®Ÿè£…ã—ã¾ã™ã€‚
  
  - ï¼ˆè¨“ç·´æ™‚ã®ã¿ï¼‰ã¾ãšã¯è¨ˆç®—ã—ã¦ã„ã‚‹ãƒŸãƒ‹ãƒãƒƒãƒã«ã¤ã„ã¦ã€å¹³å‡ã¨åˆ†æ•£ã‚’æ±‚ã‚ã¾ã™ã€‚å„æ¬¡å…ƒã«ã¤ã„ã¦ã€å…¨ãƒ‡ãƒ¼ã‚¿ã‚’é€šã˜ãŸå¹³å‡ãƒ»åˆ†æ•£ã‚’è¨ˆç®—ã™ã‚‹ãŸã‚ã€å¹³å‡ãƒ»åˆ†æ•£ã‚’è¨ˆç®—ã™ã‚‹è»¸ã«ã”æ³¨æ„ãã ã•ã„ã€‚

  - ï¼ˆè¨“ç·´æ™‚ã®ã¿ï¼‰ãƒ†ã‚¹ãƒˆæ™‚ã«ä½¿ç”¨ã™ã‚‹ãŸã‚ã«ã€è¨“ç·´ãƒ‡ãƒ¼ã‚¿å…¨ä½“ã§ã®å¹³å‡ã‚’æ¨å®šã—ã¾ã™ã€‚ãƒ¢ãƒ¼ãƒ¡ãƒ³ã‚¿ãƒ  $m$ ã‚’ç”¨ã„ã¦ä»Šã¾ã§ã®å¹³å‡ $\mu_{old} $ ã‚’è¨ˆç®—ã—ã¦ã„ã‚‹ãƒŸãƒ‹ãƒãƒƒãƒã®å¹³å‡ $\mu$ ã®æ–¹å‘ã«ç§»å‹•ã•ã›ã€æ–°ã—ã„å¹³å‡$\mu_{new} $ã‚’æ±‚ã‚ã¾ã™ã€‚
  $$
  \mu_{new} = m \mu_{old} + ( 1 - m)\mu\tag{1}
  $$

  - ï¼ˆè¨“ç·´æ™‚ã®ã¿ï¼‰åŒæ§˜ã«ä»Šã¾ã§ã®åˆ†æ•£ $\sigma_{old} ^ 2$ ã‚’è¨ˆç®—ã—ã¦ã„ã‚‹ãƒŸãƒ‹ãƒãƒƒãƒã®å¹³å‡ $\sigma^2$ã®æ–¹å‘ã«ç§»å‹•ã•ã›ã€ æ–°ã—ã„åˆ†æ•£$\sigma_{new}^2$ ã‚’æ±‚ã‚ã¾ã™ã€‚
  $$
  \sigma_{new}^2 = m \sigma_{old}^2 + ( 1 - m)\sigma^2\tag{2}
  $$

  - æ±‚ã‚ãŸå¹³å‡ $\mu$ ã¨åˆ†æ•£ $\sigma^2$ ã‚’ç”¨ã„ã¦ã€å…¥åŠ› $x$ ã‚’æ­£è¦åŒ–ã—ãŸå€¤ $x_n$ ã‚’æ±‚ã‚ã¾ã™ã€‚åˆ†æ•£$\sigma^2$ã‹ã‚‰æ¨™æº–åå·® $\sigma$ ã‚’æ±‚ã‚ã‚‹ã¨ãã«ã€ã‚¢ãƒ³ãƒ€ãƒ¼ãƒ•ãƒ­ãƒ¼ã‚’é¿ã‘ã‚‹ãŸã‚ã« 10e-7 ($10 \times 10 ^ {-7}$) ã‚’è¶³ã—ã¦ã‹ã‚‰å¹³æ–¹æ ¹ã‚’å–ã£ã¦ã„ã¾ã™ã€‚
  ãƒ†ã‚¹ãƒˆæ™‚ã«ã¯ã€ç§»å‹•å¹³å‡ã«ã‚ˆã‚Šæ¨å®šã—ãŸè¨“ç·´ãƒ‡ãƒ¼ã‚¿å…¨ä½“ã§ã®å¹³å‡ãƒ»åˆ†æ•£ã‚’ä½¿ç”¨ã—ã¾ã™ã€‚
  $$
  \sigma = \sqrt{\sigma ^ 2 + 10 \times 10 ^ {-7}}\tag{3}
  $$
  $$
  x_n = (x - \mu) / \sigma\tag{4}
  $$

   - æ­£è¦åŒ–ã—ãŸå€¤ $x_n$ã«å¯¾ã—ã¦ $\gamma$ ã‚’ç”¨ã„ã¦å¤‰å€ã—ã€$\beta$ ã‚’ç”¨ã„ã¦ç§»å‹•ã‚’è¡Œã„ã€æ´»æ€§åŒ–é–¢æ•°ã«æ¸¡ã™å‡ºåŠ› $y$ ã‚’æ±‚ã‚ã¾ã™ã€‚
   $$
   y = \gamma x_n + \beta\tag{5}
   $$
 
   ãƒãƒƒãƒæ­£è¦åŒ–ã®èª¤å·®é€†ä¼æ’­ã¯ä»¥ä¸‹ã®å¼ã«å¾“ã£ã¦å®Ÿè£…ã—ã¾ã™ã€‚
   
   - ç›´å‰ã¾ã§é€†ä¼æ’­ã—ã¦ããŸ$1, 2, \dots , N$ ç•ªç›®(Nã¯ãƒãƒƒãƒã‚µã‚¤ã‚º)ã®å‡ºåŠ›ãƒ‡ãƒ¼ã‚¿$y_k$ã«ã‚ˆã‚‹å‹¾é… $\frac{\partial L}{\partial y_k}$ã‚’ç”¨ã„ã¦ $\gamma$ ã¨$\beta$ã«ã‚ˆã‚‹å‹¾é…ã‚’è¨ˆç®—ã—ã¾ã™ã€‚ $x_{nk}$ ã¯ãƒŸãƒ‹ãƒãƒƒãƒã®ä¸­ã®kç•ªç›®ã®å…¥åŠ›ãƒ‡ãƒ¼ã‚¿ã‚’æ­£è¦åŒ–ã—ãŸå¾Œã®å€¤ã‚’è¡¨ã—ã¾ã™ã€‚
   
   $$
   \begin{eqnarray} 
   \frac{\partial L}{\partial \gamma} & = & \sum_{k=1}^{N} \frac{\partial L}{\partial y_k} \frac{\partial y_k}{\partial \gamma} =\sum_{k=1}^{N} \frac{\partial L}{\partial y_k} x_{nk}\tag{6}\\
   \frac{\partial L}{\partial \beta} & = & \sum_{k=1}^{N} \frac{\partial L}{\partial y_k} \frac{\partial y_k}{\partial \beta} =            \sum_{k=1}^{N} \frac{\partial L}{\partial y_k}\tag{7}
   \end{eqnarray}
   $$
   
   - $1, 2, \dots , N$ ç•ªç›®ã®å…¥åŠ›ãƒ‡ãƒ¼ã‚¿$x_k$ã«ã‚ˆã‚‹å‹¾é… $\frac{\partial L}{\partial x_k}$ã‚’è¨ˆç®—ã—ã¾ã™ï¼ˆã‚³ãƒ¼ãƒ‰ã§ã¯é«˜é€ŸåŒ–ã®ãŸã‚å°‘ã€…ç•°ãªã£ãŸè¨ˆç®—ã‚’ã—ã¦ã„ã¾ã™ï¼‰ã€‚
   
   $$
  \begin{equation} 
  \frac{\partial L}{\partial x_k} 
  = \frac{\gamma}{\sigma} \Bigg[ \frac{\partial L}{\partial y_k} 
  - \frac{1}{N} \bigg[ \frac{\partial L}{\partial \beta} + x_{nk} \frac{\partial L}{\partial \gamma} \bigg] \Bigg] 
  \end{equation}
  $$

```python
#é…åˆ—ãƒ¡ãƒ¢
# a = [100,100,100,100]
# b = 784
# c = 10
# d = [b] + a  + [c] 
# print(d)
# print(len(d))
```

```python
class mnistMultiLayerBatchNet:
    """
    layer0: 784 æ¬¡å…ƒã®å…¥åŠ›
    â†“ w1, b1 ã§ç·šå½¢çµåˆ
    â†“ãƒãƒƒãƒæ­£è¦åŒ– gamma1å€ã—beta1ã ã‘ãšã‚‰ã™
    â†“ relu ã§æ´»æ€§åŒ–
    layer1: 100 æ¬¡å…ƒã®éš ã‚Œå±¤
    â†“ w2, b2 ã§ç·šå½¢çµåˆ
    â†“ãƒãƒƒãƒæ­£è¦åŒ– gamma2å€ã—beta2ã ã‘ãšã‚‰ã™
    â†“ relu ã§æ´»æ€§åŒ–
    layer2: 100 æ¬¡å…ƒã®éš ã‚Œå±¤
    â†“ w3, b3 ã§ç·šå½¢çµåˆ
    â†“ãƒãƒƒãƒæ­£è¦åŒ– gamma3å€ã—beta3ã ã‘ãšã‚‰ã™
    â†“ relu ã§æ´»æ€§åŒ–
    layer3: 100 æ¬¡å…ƒã®éš ã‚Œå±¤
    â†“ w4, b4 ã§ç·šå½¢çµåˆ
    â†“ãƒãƒƒãƒæ­£è¦åŒ– gamma4å€ã—beta4ã ã‘ãšã‚‰ã™
    â†“ relu ã§æ´»æ€§åŒ–
    layer4: 100 æ¬¡å…ƒã®éš ã‚Œå±¤
    â†“ w5, b5 ã§ç·šå½¢çµåˆ
    layer5: 10 æ¬¡å…ƒã®å‡ºåŠ›å±¤
    """
    def __init__(self):
        self.input_size = 784
        self.output_size = 10
        self.hidden_size_list = [100, 100, 100, 100]
        self.all_size_list = [self.input_size] + self.hidden_size_list + [self.output_size]
        self.hidden_layer_num = len(self.hidden_size_list)
        self.weight_decay_lambda =0
        self.params = {}
        self.layers = {}
        self.grads = {}
        self.norms = {}
        self.momentum = 0.9

        # ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã®åˆæœŸåŒ–
        for idx in range(1, len(self.all_size_list)):
            # ç·šå½¢çµåˆå±¤ã®ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿
            self.params['w' + str(idx)] = np.random.randn(self.all_size_list[idx-1], self.all_size_list[idx]) * 0.085
            self.params['b' + str(idx)] = np.zeros(self.all_size_list[idx], dtype=float)
            
            # ãƒãƒƒãƒæ­£è¦åŒ–ã§ã‚·ãƒ•ãƒˆã•ã›ã‚‹ã¨ãã«ç”¨ã„ã‚‹Î³ã¨Î²ã‚’æ›´æ–°ã™ã‚‹ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã¨ã—åˆæœŸåŒ–
            # mu ã¨ sigma ã¯å®Ÿè¡Œæ™‚ã®å¹³å‡ã¨åˆ†æ•£
            if idx != len(self.all_size_list) - 1:
                self.params['gamma' + str(idx)] = np.ones(self.all_size_list[idx])
                self.params['beta' + str(idx)] = np.zeros(self.all_size_list[idx])
                self.norms['mu' + str(idx)] = None
                self.norms['var' + str(idx)] = None
        
    def forward(self, x, train_flg=False):
        relu = lambda x : np.maximum(0, x)  # æ´»æ€§åŒ–é–¢æ•°ã¨ã—ã¦ ReLU ã‚’ä½¿ç”¨
        self.layers['layer0'] = x
        for idx in range(1, len(self.all_size_list) - 1):
            # ç·šå½¢çµåˆå±¤
            w = self.params['w' + str(idx)]
            b = self.params['b' + str(idx)]
            x = self.layers['layer' + str(idx - 1)]
            x = np.dot(x, w) + b
            
            # ãƒãƒƒãƒæ­£è¦åŒ–
            # å¹³å‡ã¨åˆ†æ•£ã®åˆæœŸåŒ–
            if self.norms['mu' + str(idx)] is None:
                N, D = x.shape
                self.norms['mu' + str(idx)] = np.zeros(D)
                self.norms['var' + str(idx)] = np.zeros(D)
            if train_flg:
                ###### å•2.1 ######
                #ãƒãƒƒãƒã”ã¨ãªã‚‰ã°axisã¯1ã«ãªã‚‹ï¼Ÿ
                #mu = np.mean(x,axis = 1)          # ä»Šå›ã®ãƒŸãƒ‹ãƒãƒƒãƒã®å¹³å‡
                mu = np.mean(x,axis = 0)
                xc = x - mu                   # ä»Šå›ã®ãƒŸãƒ‹ãƒãƒƒãƒã®å¹³å‡ã¨ã®å·®åˆ†
                ###### å•2.2 ######
                #var = np.var(x,axis = 1)  # ä»Šå›ã®ãƒŸãƒ‹ãƒãƒƒãƒã®åˆ†æ•£
                var = np.var(xc,axis = 0)
                std = np.sqrt(var + 10e-7)    # ä»Šå›ã®ãƒŸãƒ‹ãƒãƒƒãƒã®æ¨™æº–åå·®
                xn = xc / std                 # æ­£è¦åŒ–

                # å…¨ä½“ã®å¹³å‡ã¨åˆ†æ•£ã‚’ç§»å‹•å¹³å‡ã«ã‚ˆã‚Šæ±‚ã‚ã‚‹(1),(2)
                ###### å•2.3 ######
                #å¹³å‡oldã®è¨˜è¼‰ã¯ã“ã‚Œã§ã‚ã£ã¦ã„ã‚‹ï¼Ÿidx=1ã®ã¨ãã¯
                self.norms['mu' + str(idx)] = self.momentum * self.norms['mu' + str(idx)] + (1 - self.momentum) * mu
                ###### å•2.4 ######
                self.norms['var' + str(idx)] = self.momentum * self.norms['var' + str(idx)] + (1 - self.momentum) * var
                
                # èª¤å·®é€†ä¼æ’­ã§ä½¿ã†ä¸­é–“ãƒ‡ãƒ¼ã‚¿
                self.norms['xc' + str(idx)] = xc
                self.norms['xn' + str(idx)] = xn
                self.norms['std' + str(idx)] = std
                self.norms['size' + str(idx)] = x.shape[0]
            else:
                # ãƒ†ã‚¹ãƒˆæ™‚ã¯å…¨ä½“ã®å¹³å‡ã¨åˆ†æ•£ã‚’ä½¿ã£ã¦æ­£è¦åŒ–ã™ã‚‹(3),(4)
                ###### å•2.5 ######
                #xc = np.sqrt(np.var(x, axis = 1) + 10e-7)
                xc = x - self.norms['mu'+str(idx)]
                ###### å•2.6 ######
                #xn = (x - np.mean(x , axis = 1) ) / xc
                xn = xc / np.sqrt(self.norms['var'+str(idx)] + 10e-7)
                
            # ãƒãƒƒãƒæ­£è¦åŒ–ã§ã‚·ãƒ•ãƒˆã•ã›ã‚‹(5)
            ###### å•2.7 ######
            #ğ‘¦=ğ›¾*ğ‘¥_ğ‘›+ğ›½  ğ›¾:æ¨™æº–åå·®ï¼ˆå„æ¬¡å…ƒï¼‰ ğ›½:å¹³å‡ã¨ã„ã†èªè­˜ï¼ˆå„æ¬¡å…ƒï¼‰
            #shifted = np.sqrt(np.var(x, axis = 0) + 10e-7)*xn + np.mean(x , axis = 0)
            shifted = self.params['gamma' + str(idx)]*xn +self.params['beta'+str(idx)]

            # relu ã‚’ä½¿ã£ã¦æ´»æ€§åŒ–
            self.layers['layer' + str(idx)] = relu(shifted)

        # å‡ºåŠ›å±¤
        idx = len(self.all_size_list) - 1
        w = self.params['w' + str(idx)]
        b = self.params['b' + str(idx)]
        x = self.layers['layer' + str(idx - 1)]
        self.layers['layer' + str(idx)] = softmax(np.dot(x, w) + b)
        
        return self.layers['layer' + str(idx)]
        

    def loss(self, y, t):
        return cross_entropy_error(y, t)
    
    def backward(self, t, y):
        # å‡ºåŠ›å±¤ã«ãŠã‘ã‚‹èª¤å·®ã®å‹¾é…ï¼ˆã‚¯ãƒ­ã‚¹ã‚¨ãƒ³ãƒˆãƒ­ãƒ”ãƒ¼é–¢æ•°ã®å‹¾é…ï¼‰
        delta = (y - t) / t.shape[0]
        
        # å‡ºåŠ›å±¤æ‰‹å‰ã®ç·šå½¢çµåˆå±¤ã«ãŠã‘ã‚‹å‹¾é…ã®é€†ä¼æ’­
        self.grads['b5'] = np.sum(delta, axis=0)
        self.grads['w5'] = np.dot(self.layers['layer4'].transpose(), delta)
        
        # èª¤å·®é€†ä¼æ’­
        for idx in range(4, 0, -1):
            delta = np.dot(delta, self.params['w' + str(idx + 1)].transpose())
            
            # relu ã®å¾®åˆ†
            delta = delta *  (self.layers['layer' + str(idx)] > 0)
            
            # ãƒãƒƒãƒæ­£è¦åŒ–ã«ãŠã‘ã‚‹å‹¾é…ã®é€†ä¼æ’­(6),(7)
            ###### å•2.8 ######
            self.grads['beta' + str(idx)] = np.sum(delta, axis=0) 
            ###### å•2.9 ######
            self.grads['gamma' + str(idx)] = np.sum(delta * self.norms['xn' + str(idx)] , axis = 0)
            dxn = self.params['gamma' + str(idx)] * delta
            dxc = dxn / self.norms['std' + str(idx)]
            dstd = -np.sum((dxn * self.norms['xc' + str(idx)]) / (self.norms['std' + str(idx)] * self.norms['std' + str(idx)]), axis=0)
            dvar = 0.5 * dstd / self.norms['std' + str(idx)]
            dxc += (2.0 / self.norms['size' + str(idx)]) * self.norms['xc' + str(idx)] * dvar
            dmu = np.sum(dxc, axis=0)
            delta = dxc - dmu / self.norms['size' + str(idx)]
            
            # ç·šå½¢çµåˆå±¤ã«ãŠã‘ã‚‹å‹¾é…ã®é€†ä¼æ’­
            self.grads['b' + str(idx)] = np.sum(delta, axis=0)
            self.grads['w' + str(idx)] = np.dot(self.layers['layer'+str(idx - 1)].transpose(), delta)
            
        return self.grads
```
